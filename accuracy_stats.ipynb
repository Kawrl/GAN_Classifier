```python
# Plot grid with four images from each class:
import numpy as np
import torchvision

def plot_samples(dataset,fname):   
    random_idxs = np.random.randint(low=0,high=len(dataset),size=16)

    imgs = []
    for idx in random_idxs:
        img,label,_ = dataset[idx]
        imgs.append(img)
    grid=torchvision.utils.make_grid(imgs,normalize=False)
    mean=0.5
    std=0.5
    grid = (grid + (mean/std))/(1/std)
    fig, ax = plt.subplots(figsize=(16,16))
    ax.imshow(grid.permute(1, 2, 0))
    ax.set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    plt.savefig(fname+'.png',bbox_inches='tight')
```

```python
plot_samples(acgan200,'16samples_acgan200')
plot_samples(acgan1000,'16samples_acgan1000')
plot_samples(reacgan200,'16samples_reacgan200')
plot_samples(reacgan1000,'16samples_reacgan1000')
```
